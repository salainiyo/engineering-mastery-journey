```python
import pandas as pd
import numpy as np
```

```python
try:
    data = pd.read_csv("data/synthetic_shopping_data.csv")
except Exception as e:
    print(str(e))
```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
categorical_data = data.select_dtypes(include=['object', 'category', 'str']).columns.tolist()
categorical_data
```

```python
numerical_data = data.select_dtypes(include=['int', 'float']).columns.to_list()
numerical_data
```

```python
for numerical_column in numerical_data:
    data[numerical_column] = data[numerical_column].fillna(data[numerical_column].mean())
for cat_column in categorical_data:
    data[cat_column] = data[cat_column].fillna(data[cat_column].mode()[0])

data.isnull().sum()
```

```python
data = data.drop(['id'], axis=1)
data.isnull().sum()
```
